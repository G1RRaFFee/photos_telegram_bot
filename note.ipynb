```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
file_path = 'Consumption of alcoholic beverages in Russia 2017-2023.csv'
data = pd.read_csv(file_path)

# Проверка пропущенных данных
print("Пропущенные значения:\n", data.isnull().sum())

# Статистическая информация
print("\nОсновные статистики:\n", data.describe())

# Среднее потребление алкоголя по годам
avg_consumption_per_year = data.groupby('Year')['Total alcohol consumption (in liters of pure alcohol per capita)'].mean()

# Визуализация: среднее потребление алкоголя по годам
plt.figure(figsize=(10, 6))
sns.lineplot(x=avg_consumption_per_year.index, y=avg_consumption_per_year.values, marker="o", color="blue")
plt.title("Среднее потребление алкоголя (литры на душу) по годам", fontsize=14)
plt.xlabel("Год", fontsize=12)
plt.ylabel("Среднее потребление алкоголя", fontsize=12)
plt.grid(visible=True)
plt.show()

# Гипотеза: корреляция между потреблением водки и общим потреблением алкоголя
correlation = data['Vodka'].corr(data['Total alcohol consumption (in liters of pure alcohol per capita)'])
print("\nКорреляция между потреблением водки и общим потреблением алкоголя:", correlation)

# Визуализация: зависимость между потреблением водки и общим объемом потребления
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=data['Vodka'], 
    y=data['Total alcohol consumption (in liters of pure alcohol per capita)'], 
    alpha=0.7, color="red"
)
plt.title("Корреляция между потреблением водки и общим потреблением алкоголя", fontsize=14)
plt.xlabel("Потребление водки (литры на душу)", fontsize=12)
plt.ylabel("Общее потребление алкоголя (литры на душу)", fontsize=12)
plt.grid(visible=True)
plt.show()
```
